```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({'A':[1,2,np.nan],
                   'B':[5,np.nan,np.nan],
                   'C':[1,2,3]})
df
```

```python
#fillna()
df = pd.DataFrame({'A':[1,2,np.nan],
                   'B':[5,np.nan,np.nan],
                   'C':[1,2,3]})
df
```

```python
df.fillna(value=4.5)
```

```python
#groupby
df = pd.DataFrame({'Company':['GOOG','GOOG','MSFT','MSFT','FB'],
                   'Person':['Sam','Charlie','Amy','Vanessa','komal',],
                   'Salary':[50000,850000,70000,95120,956210,]})
df
```

```python
#select count(*) from emp group by company
df.groupby('Company').count()
```

```python
df.groupby('Company').mean('Salary')
```

```python
df.groupby('Company').describe()
```

```python
df.groupby('Company').describe().transpose()
```
