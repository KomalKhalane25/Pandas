```python
#merge()

df1 = pd.DataFrame({'Id':[1,2,3,4],
                    'Name':['komal','Rahul','Rani','Om'],
                    'sub_Id':['sub1','sub2','sub3','sub4']
                    })
df2 = pd.DataFrame({ 'Id':[1,2,3,4],
                    'Name':['Minal','Prajol','Rani','Omkar'],
                    'sub_Id':['sub6','sub5','sub3','sub4'],
                    })
df3 = pd.DataFrame({ 'Id':[1,2,33,44],
                    'Name':['Minal','Raj','Rani','Omkar'],
                    'sub_Id':['sub1','sub2','sub3','sub4'],
                    })

print(df1)
print('***********************')
print(df2)
print('***********************')
print(df3)
```

```python
pd.merge(df1,df3,on='Id',how='inner')
```

```python
pd.merge(df1,df3,on='Id',how='left')
```

```python
pd.merge(df1,df3,on='Id',how='right')
```

```python
pd.merge(df1,df3,on='Id',how='outer')
```

```python
print(df1)
print('***********************')
print(df3)
pd.merge(df1,df3,on=['Id','sub_Id'])
```

```python
x = pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3']},
                  index=['K0','K1','K2','K3'])
y = pd.DataFrame({'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},
                  index=['K0','K1','K2','K3'])

print(x)
print('*************')
print(y)
x.join(y)  #by default it is left join
```

```python
print(x)
print('*************')
print(y)
x.join(y,how ='right')
```

```python
print(x)
print('*************')
print(y)
x.join(y,how ='outer')
```
