```python
#Pandas Series
```

```python
import numpy as np
import pandas as pd
```

```python
arr = np.array([1,2,3,4])
print(arr)
```

```python
ser = pd.Series(arr)
print(ser)
```

```python
ser[1]
```

```python
ser1 = pd.Series([1,2,3,4],index = ['India','USA','Germany','Japan'])
print(ser1)
```

```python
ser1['India']
```

```python
ser2 = pd.Series(arr,index = ['a','b','c','d'])
print(ser2)
```

```python
ser3 = pd.Series({'a':1.5,'b':'komal',})
print(ser3)
```

```python
ser3[0]
```

```python
ser1 = pd.Series([10,20,30],index=[1,2,3])
ser2 = pd.Series([20,40,60],index=[1,3,2])
print(ser1+ser2)
```

```python
ser1 = pd.Series([10,20,30],index=[1,2,3])
ser2 = pd.Series([20,40,60,50],index=[1,3,2,4])
print(ser1+ser2)
```

```python
ser1 = pd.Series([10,20,30,50],index=[0,1,2,3])
ser2 = pd.Series([20,40,60],index=[1,3,2])
print(ser1+ser2)
```

```python
print(ser1-ser2)
```

```python
print(ser1*ser2)
```
